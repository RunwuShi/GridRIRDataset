```python
import os
import json
import soundfile as sf
import numpy as np
from natsort import natsorted 

# --------------------------------------------------------------------------
# W504_01 info 
# The mic is in the center of the room
# Each point is measured four times, up, right, down, left #
# Channel 1 is negative direction, direction of gate to window is positive.
# --------------------------------------------------------------------------
# W504_02, I have only used the RIR from this one
# The mic is in the center of the room
# Each point is measured once
# 05 06 Missing center position because mic and loudspeaker are too close together
# mic height 1.0m
# --------------------------------------------------------------------------
# W504_03
# mic is in the upper right corner of the room
# --------------------------------------------------------------------------
# Take the first position of the first row in the lower left corner as the relative center
# Center mic position: [2m, 2.5m]
# --------------------------------------------------------------------------
# data folder path
full_root_path = ['./grid02']
# -> save path
save_path = './grid02_info'
# --------------------------------------------------------------------------

# from top to down
# full position
row_posi_list = [4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0.5, 0.]
column_posi_list = [0., 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5]
# lack of middle position
column_posi_list_2 = [0., 0.5, 1.0, 1.5, 2.5, 3.0, 3.5, 4.0, 4.5]

# mic posi
mic_posi = [2.0, 2.5, 1.0]

# room dim
room_dim = [5.9, 6.9, 2.9]

# total distance
total_dis_2d = []

# total data
total_speaker_num = 98
total_mic_num = 8
total_room_num = 1
total_room_info = [[[[] for _ in range(total_speaker_num)] for _ in range(total_mic_num)] for i in range(total_room_num)]

room_index = 0
for root_path in full_root_path:
    spk_index = 0
    
    # raw first
    row_index = 0
    root_paths = [f for f in os.listdir(root_path) if not f.startswith('.')]
    sort_root_paths = natsorted(root_paths)
    print('sort_root_paths', sort_root_paths)
    
    for folder_name in sort_root_paths:
        print('folder_name', folder_name)
        
        folder_path = os.path.join(root_path, folder_name)
        one_file_paths = [f for f in os.listdir(folder_path) if not f.startswith('.')]
        sort_one_file_paths = natsorted(one_file_paths)
        # print('one_file_path', one_file_path)
        # print(sorted_posis_per_row)
        # print(len(sorted_posis_per_row))
        
        # column
        column_index = 0
        for one_posi in sort_one_file_paths:
            # print(sort_one_file_paths)
            # print(len(sort_one_file_paths))
            
            if len(sort_one_file_paths) == 10:
                row_posi = row_posi_list[row_index]
                column_posi = column_posi_list[column_index]
                
            elif len(sort_one_file_paths) == 9:
                row_posi = row_posi_list[row_index]
                column_posi = column_posi_list_2[column_index]
                
            # print(one_posi)
            # w504->02->mic01_01->newrir1

            # single RIR file
            RIR_folder_path = os.path.join(folder_path, sort_one_file_paths[column_index])
            
            # src position
            # height of the mic: 0.9, Table height: 88cm, flare 100cm
            src_posi = [column_posi, row_posi, 0.9] 
            
            # mic position
            mic_posi = mic_posi
            
            # distance
            distance_2d = np.sqrt((src_posi[0]-mic_posi[0])**2 + (src_posi[1]-mic_posi[1])**2)
            distance_3d = np.sqrt((src_posi[0]-mic_posi[0])**2 + (src_posi[1]-mic_posi[1])**2 + (src_posi[2]-mic_posi[2])**2)
            
            total_dis_2d.append(distance_2d)
            
            # room dim
            room_dim = room_dim 
            
            # RIR files
            RIR_names = ['RIR{}.wav'.format(i) for i in range(1, 9)]
            
            # for all mics
            for mic_index, RIR_name in enumerate(RIR_names):
                
                RIR_path = os.path.join(RIR_folder_path, RIR_name)
                # print('RIR_folder_path',RIR_folder_path)
                # print('src_posi', src_posi)
                # print('RIR_path', RIR_path)

                # save info
                info = {
                    "position_index": spk_index,
                    "rever_path": RIR_path,
                    "room_dim": room_dim,
                    "source_posi": src_posi,
                    "mic_posi": mic_posi,
                    "dis_to_mic_2D": round(distance_2d, 3),
                    "dis_to_mic_3D": round(distance_3d, 3)
                }

                print('spk_index',spk_index)
                total_room_info[room_index][mic_index][spk_index].append(info)

            column_index += 1
            spk_index += 1
            
        row_index += 1
        

os.makedirs(save_path, exist_ok=True)
file_path = os.path.join(save_path, 'full_total_data_info.json')

# save as json
save = True
if save:
    with open(file_path, 'w') as f:
        json.dump(total_room_info, f, indent=4)
```
